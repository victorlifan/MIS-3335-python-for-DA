# Basic Python data types
What is a data type?
- “… classification of data which tells the compiler or interpreter how the programmer intends to use the data.” (Wikipedia, 2017)
- We've already seen the difference between text and numbers.
   - We have different number types, too: integers versus decimal values.

There are relatively few data types in Python
- Boolean: True or False
- int: any whole number
- float: any decimal value
- str: any alphanumeric or symbol text
- datetime: any time or date value
- list: multiple value structure of any other data type
- tuple: multiple value structure of any other data type that can’t be changed
- dictionary: key:value pairs

Remember using quotation marks for string literals but not for numeric literals?
- This is how Python determines if a variable contains a number or a string.

This is important because in order to “do math” on a variable, it must be a numeric data type.

## *Variables can vary.*



That's where the name comes from. The value stored in a variable can change, which makes our code more flexible. But sometimes we want the values in a named memory location to NOT change, to remain constant. We call those memory locations ... wait for it ... `constants`... and we can give them names, too. **_Named constants_** use all caps and underscores between words.

A number written into code (“hard coded”) is called a numeric literal. **_We want to avoid numeric literals at all costs._** Use named constants instead. We will also put those statements at the beginning of our notebook so they are easy to find.

````python
# numeric literal in calculation - BAD PRACTICE
subtotal = 1000   # could be the result of an earlier calculation
sales_tax_amount = subtotal * 0.0885
````

````python
# named constant in calculation - GOOD PRACTICE
SALES_TAX_RATE = 0.0885
subtotal = 1000
sales_tax_amount = subtotal * SALES_TAX_RATE
````

## Lists
A list can hold data of the same type or of different types. Each item in a list is called an _**element**_. If you are familiar with another programming language, it is similar to an array. Not _exactly_ the same as an array, but the differences are technical in nature and don't make any difference as far as we are concerned.

We can create a list by assigning elements separated by commas inside square brackets [] to the list name. List names follow the same rules as variable names.

````python
# Create a list of movie titles
movies = ['Scarface', 'The Blob', 'Star Wars', 'Heidi']
````

We use _indexing_ to refer to the individual elements in a list. The **index** specifies the position of the desired element in the list, beginning with _the first element at position 0 (zero)_. As we have seen before this is referred to as a **zero-based index**. It is one concept that beginning programmers have a hard time with because it is different than what we are used to. One implication is that the last index in a list and the length of the list are not the same!

When we want a specific list element, we specify the index inside the square brackets

````python
# display the element at index 2
print(movies[2])
````

## Dictionaries
With a Python list, we have to access elements in sequence. That means the order in which the elements appear in the list might have meaning. That's fine if that is helpful, but what if you didn't care about the order of the elements because you had a way to access them directly? That is what the Python dictionary allows you to do.

What is a **_dictionary_**?
- A dictionary is an object that stores a collection of data.
- Each element has two parts: a **key** and a **value**.
- _You use a key to find a specific value. You do not use an index._
- The keys have to be immutable objects like strings, integers, floating-point values or tuples.  
- Keys also have to be unique – **no duplicate keys**.
- *Values can be objects of any type*, so they can be any of those objects above plus lists.

Dictionaries are created using the curly braces `{}`. One way to create and populate a dictionary is to use the curly braces to enclose a listing of the elements. Separate the key and the value with a colon. Strings require quotes but numeric & Boolean values do not.


````python
# Create two dictionaries
employee = {1:'John', 23:'Eric', 38:'Graham', 4:'Terry', 50:'Terry'}
offices = {'John':'COB 201', 'Eric':'Wingo 107', 'Graham':'COB 305'}
````

Use the dictionary name and the key in square brackets to retrieve a value. String keys are case sensitive. They must match exactly.

````python
# Access dictionary values using the key
print(employee[38], '     ',offices['Eric'])
````
