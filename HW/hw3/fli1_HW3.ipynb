###### SETUPS

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# read in datasets
df_apple=pd.read_csv("Applewood_2011.csv")
df_college=pd.read_csv("College.csv")
df_housing=pd.read_csv("Boston_housing.csv",sep=";")
```

###### Dataset 1 – Applewood_2011.csv

```python
df_apple
```

1. Show all the pairwise plots.

```python
sns.pairplot(df_apple);
```

2. Show a bar chart of the number of sales by Location.

```python
sns.countplot(data=df_apple,x='Location',color='c');
```

3. Show box plots for profit with each value of “Previous” as a separate box. 

```python
sns.boxplot(data=df_apple,x='Previous',y='Profit',color='c');
```

4. Scatter plots with each value of Previous on its own plot, three across

```python
g=sns.FacetGrid(data=df_apple,col='Previous',col_wrap=3)
g.map(plt.scatter,'Age','Profit');
```

```python
g=sns.FacetGrid(data=df_college,col='Private',height=5)
g.map(sns.regplot,'Outstate',"Room.Board");
```

###### Dataset 3 – Boston_housing.csv

```python
df_housing
```

1. Scatter plot of LSTAT and MEDV, with the value of CHAS on separate plots

```python
g=sns.FacetGrid(data=df_housing,col='CHAS',height=5)
g.map(plt.scatter,"LSTAT","MEDV");
```

2. Box plot of median value with CHAS as the x variable. 

```python
sns.boxplot(data=df_housing,x="CHAS",y="MEDV",color='c')
plt.title("Median Value by Charles River Proximity")
plt.xlabel('Near the Charles River?')
plt.ylabel("Median Home Value");
```

3. Scatter plot without regression line of rooms versus median value

```python
sns.lmplot(data=df_housing,x="RM",y="MEDV",fit_reg=False)
plt.title("Number of Rooms versus Median Value")
plt.xlabel("Number of Rooms")
plt.ylabel("Median House Value (in thoustands)");
```
