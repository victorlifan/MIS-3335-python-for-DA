```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# read in datasets and replace NaN
dom=pd.read_csv("lhr_visit_dom.csv",na_values=['null','none'])
intl=pd.read_csv("lhr_visit_intl.csv",na_values=['null','none'])
```

```python
dom
```

```python
intl.info()
```

```python
# Combine the two dataframes into a new dataframe called “lhr_visits.”
lhr_visits=pd.concat([dom,intl],ignore_index=True)
lhr_visits
```

```python
# Fill all missing values with a zero value
lhr_visits.fillna(0,inplace=True)
lhr_visits
```

```python
# Save the data in a csv file called “lhr_visits.csv”
lhr_visits.to_csv("lhr_visits.csv",index=False)
```

```python
# Load the data from “lhr_visits.csv” into a dataframe called “visits.”
visits=pd.read_csv("lhr_visits.csv")
visits
```

Melt the data columns into a column called “visit_length” and a column called “count.”

```python
visits=pd.melt(visits,id_vars='type',var_name='visit_length',value_name='count')
visits
```

Split the values in the “type” column into values for variables called “residence”, 
“purpose”, and “flight_type.” Add those three columns to the dataframe.

```python
visits=visits['type'].str.split('_',expand=True).rename(columns={0:'residence',1:'purpose',2:'flight_type'}).join(visits)
visits
```

```python
# Drop the “type” column
visits.drop('type',axis=1)
```
