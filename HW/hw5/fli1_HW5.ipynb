```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# read in datasets with the ID column set to be the index value
fifa=pd.read_csv("../../data/fifa19_ver1.csv",index_col='ID')
fifa.head()
```

###### Part 1

```python
fifa.dtypes
```

```python
# Photo
fifa['player_photo']=fifa['Photo'].str.split("https://cdn.sofifa.org/players/4/19/").str[1]
fifa.drop('Photo',axis=1,inplace=True)
```

```python
fifa.head()
```

```python
# Wage
fifa['wage_euros']=fifa['Wage'].replace({'€':"",'K':""},regex=True).str.strip().astype(float)
fifa.drop('Wage',axis=1,inplace=True)
fifa.head()
```

```python
fifa.dtypes
```

```python
# category columns
fifa[list(fifa)[7:11]]=fifa[list(fifa)[7:11]].astype("category")
fifa.dtypes
```

```python
# Jersey Number
fifa['Jersey Number'].fillna(0,inplace=True)
fifa['Jersey Number']=fifa['Jersey Number'].astype(int).astype(str)
```

```python
fifa.head()
```

```python
fifa.info()
```

###### Part 2

```python
# 1. How many players are left footed?
print(fifa.groupby('Preferred Foot')['Preferred Foot'].count(),'\n',
     "4211 players are left footed")
```

```python
# 2. Which clubs have the three highest weekly payroll totals?
print(fifa.groupby('Club')['wage_euros'].sum().nlargest(3),'\n',
     "Clubs have the three highest weekly payroll totals are: Real Madrid, FC Barcelona, Manchester City")
      
```

```python
# 3. What are the five most common jersey numbers?
print(fifa.groupby('Jersey Number')['Jersey Number'].count().nlargest(5),'\n',
     "The five most common jersey numbers are 8,7,10,11,6")
```

```python
# 4. Players from which country (i.e., “Nationality”) have the lowest average age?
print(fifa.groupby('Nationality')['Age'].mean().nsmallest(1),'\n',
      "Players from Indonesia have the lowest average age")
```
